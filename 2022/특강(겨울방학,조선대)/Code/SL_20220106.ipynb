## 함수
- 입력값을 가지고 어떤 일을 수행한 다음에 그 결과물을 내어놓는 역활을 하는 것

```python
# 더하기 함수
def cal_plus(a, b): 
  return a + b
c = cal_plus(1,3)
print(c)

# 빼기함수
def cal_minus(a, b): 
  return a - b
c = cal_minus(3,2)
print(c)

# 곱하기 함수
def cal_multiply(a, b): 
  return a * b
c = cal_multiply(3,2)
print(c)

# 나누기 함수
def cal_divide(a, b): 
  return a / b
c = cal_divide(4,2)
print(c)

# 생각하기: 더하기 함수인데 함수 내에서 지정된 값(i.e., 10)을 더 더하는 함수 만들기
# c = cal_plus(1,2)
# print(c)
# 13

def cal_plus(a, b): 
  d = 10
  return a + b + d
c = cal_plus(1,2)
print(c)


def cal_plus(a, b): 
  outcome = a + b + 10
  return outcome
c = cal_plus(1,2)
print(c)
```

## 클래스
- 함수들의 각 작업을 모아서 관리한다.
- 함수들이 동작하면서 변하는 값들을 기억하여 사용할 수 있다.

```python
# 함수들의 각 작업을 모아서 관리한다.

class Cal_class:
  def __init__(self, a, b):
    self.a = a
    self.b = b
  def cal_plus(self): 
    return self.a + self.b

  def cal_minus(self): 
    return self.a - self.b

  def cal_multiply(self): 
    return self.a * self.b

  def cal_divide(self): 
    return self.a / self.b
```

```python
a = 15
b = 5
cal = Cal_class(a, b)

print(cal.cal_plus())
print(cal.cal_minus())
print(cal.cal_multiply())
print(cal.cal_divide())
```

```python
#생각하기: 제곱을 하는 함수를 클래스에 추가하기

class Cal_class:
  def __init__(self, a, b):
    self.a = a
    self.b = b
  def cal_plus(self): 
    return self.a + self.b

  def cal_minus(self): 
    return self.a - self.b

  def cal_multiply(self): 
    return self.a * self.b

  def cal_divide(self): 
    return self.a / self.b
  
  def cal_square(self): 
    return self.a ** self.b
```

```python
a = 2
b = 3
cal = Cal_class(a, b)

print(cal.cal_square())
```

```python
# 함수들이 동작하면서 변하는 값들을 기억하여 사용할 수 있다.
class Cal_class:
  def __init__(self):
    self.result = 0

  def cal_plus(self, a): 
    self.result = self.result + a #self.result += a
    return self.result

  def cal_minus(self, a): 
    self.result = self.result - a #self.result -= a
    return self.result

  def cal_multiply(self, a): 
    self.result = self.result * a #self.result *= a
    return self.result

  def cal_divide(self, a): 
    self.result = self.result / a #self.result /= a
    return self.result
```

```python
cal = Cal_class()

print(cal.cal_plus(3))
print(cal.cal_plus(7))
print(cal.cal_minus(2))
print(cal.cal_divide(4))
print(cal.cal_multiply(3))
```
