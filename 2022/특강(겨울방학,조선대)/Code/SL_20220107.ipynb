```python
# 메타 문자(meta characters): . ^ $ * + ? { } [ ] \ | ( )
# 원래 그 문자가 가진 뜻이 아닌 특별한 용도로 사용하는 문자
import re

# 1) 정규 표현식 문법

# .기호 # a와 c 사이에는 어떤 1개의 문자라도 올 수 있다.
r = re.compile("a.c")
print(r.search("kkk")) # 아무런 결과도 출력되지 않는다.
print(r.search("abc"))
print(r.search("akc"))


# ?기호 # ?앞의 문자가 존재할 수도 있고 존재하지 않을 수도 있는 경우를 나타낸다.

r = re.compile("ab?c")
r.search("abbc") # 아무런 결과도 출력되지 않는다.
r.search("abc")
r.search("ac")

# *기호 # *은 바로 앞의 문자가 0개 이상일 경우를 나타낸다.

r = re.compile("ab*c")
r.search("a") # 아무런 결과도 출력되지 않는다.
r.search("ac")
r.search("abc") 
r.search("abbbbc") 

# +기호 # 앞의 문자가 최소 1개 이상이어야 출력한다.
r = re.compile("ab+c")
r.search("ac") # 아무런 결과도 출력되지 않는다.
r.search("abc") 
r.search("abbbbc") 

# ^기호 #시작되는 문자열을 지정한다.

r = re.compile("^ab")

r.search("bbc") # 아무런 결과도 출력되지 않는다.
r.search("zab") # 아무런 결과도 출력되지 않는다.
r.search("abz")
r.search("abzsdsfs")

# {숫자} 기호 # 해당 문자를 숫자만큼 반복한 것을 나타낸다.

r = re.compile("ab{2}c")

r.search("ac") # 아무런 결과도 출력되지 않는다.
r.search("abc") # 아무런 결과도 출력되지 않는다.
r.search("abbbbbc") # 아무런 결과도 출력되지 않는다.
r.search("abbc")

# {숫자1, 숫자2} 기호 # 해당 문자를 숫자1 이상 숫자2 이하만큼 반복하였을때 출력한다.

r = re.compile("ab{2,8}c")

r.search("ac") # 아무런 결과도 출력되지 않는다.
r.search("abc") # 아무런 결과도 출력되지 않는다.
r.search("abbbbbbbbbc") # 아무런 결과도 출력되지 않는다.

r.search("abbc")
r.search("abbbbbbbbc")

# {숫자,} 기호 # 해당 문자를 숫자 이상 만큼 반복한다.

r = re.compile("a{2,}bc")

r.search("abc") # 아무런 결과도 출력되지 않는다.

r.search("aabc")
r.search("aaaaaaaabc")

# [ ] 기호 # [ ]안에 입력된 문자들 중 한 개의 문자와 매치한다.

r = re.compile("[a-c]+") # [abc]는 [a-c]와 같다.
print(r.search("zzz")) # 아무런 결과도 출력되지 않는다.

print(r.search("baac"))

# [^문자] 기호 # ^기호 뒤에 붙은 문자들을 제외한 모든 문자를 매치하는 역할을 한다.

r = re.compile("[^abc]+")

print(r.search("zzz")) 

print(r.search("baac"))

r.search("a") # 아무런 결과도 출력되지 않는다.
r.search("ab") # 아무런 결과도 출력되지 않는다.
r.search("b") # 아무런 결과도 출력되지 않는다.

r.search("d")


# r = re.compile("[Ccats]+")

r = re.compile("cats")
r = re.compile("cat")
r = re.compile("Cats")
r = re.compile("Cat")
print("cat")
print(r.search("cat"))
print("cats")
print(r.search("cats"))
print("Cat")
print(r.search("Cat"))
print("Cats")
print(r.search("Cats"))

# 2) 정규 표현식 모듈 함수
# re.search() # 정규 표현식 전체에 대해서 문자열이 매치하는지를 검색한다.
r = re.compile("ab.")
print(r.search("kkkabc"))
print(r.search("abckkk")) 

# re.match() # 문자열의 첫 부분부터 정규 표현식과 매치하는지를 확인한다.
r = re.compile("ab.")
print(r.match("kkkabc"))
print(r.match("abckkk"))

# re.split() #입력된 정규 표현식을 기준으로 문자열들을 분리하여 리스트로 리턴한다.

# 공백 기준 분리
text = "사과 딸기 수박 메론 바나나"
re.split("\s", text)

# '+'를 기준으로 분리
text = "사과+딸기+수박+메론+바나나"
re.split("\+", text)

# re.findall() # 정규 표현식과 매치되는 모든 문자열들을 리스트로 리턴한다.

text = """이름 : 김철수 전화번호 : 010 - 1234 - 1234 나이 : 30 성별 : 남"""
re.findall("\d+", text)

# re.sub() # 정규 표현식 패턴과 일치하는 문자열을 찾아 다른 문자열로 대체한다.
import re
text = "이름 : 김철수 전화번호 : 010 - 1234 - 1234 나이 : 30 성별 : 남"

preprocessed_text = re.sub('[^가-힣]', ' ', text)
print(preprocessed_text)

# 생각하기: 위의 출력 결과에서 연속 되는 공백을 하나의 공백으로 변환하기
preprocessed_text = re.sub('[\s]+', ' ', preprocessed_text)
print(preprocessed_text)

# group() # 해당 그룹과 일치하는 문자열을 얻어올 때 사용한다.

m = re.match('([0-9]+) ([0-9]+)', '10 298298395')
print(m.group(1))    # 첫 번째 그룹(그룹 1)에 매칭된 문자열을 반환
print(m.group(2))    # 두 번째 그룹(그룹 2)에 매칭된 문자열을 반환
print(m.group())     # 매칭된 문자열을 한꺼번에 반환
print(m.group(0))    # 매칭된 문자열을 한꺼번에 반환

# # 그룹 이름 만들기

m = re.match('(?P<fd>[0-9]+) (?P<second>[0-9]+)', '10 295')
print(m.group('fd'))    # 첫 번째 그룹(first)에 매칭된 문자열을 반환
print(m.group('second'))   # 두 번째 그룹(second)에 매칭된 문자열을 반환

# 찾은 문자열을 결과에 다시 사용하기
a = 'Chosun Hello'
print(a)
a_changed = re.sub('([a-zA-Z]+) ([a-zA-Z]+)', '\\2 \\1', a)    # 그룹 2, 1, 2, 1 순으로 바꿈
print(a_changed)

```

## Bag of words (i.e., word type)

```python
Sentences = ["연어/NNG 초밥/NNG 맛있/VA 어요/EFN",
             "연인/NNG 탕수육/NNG 선택/NNG 구름/NNG 탕수육/NNG 으로/JKM 곱빼기/NNG 추가/NNG 선택/NNG 맛있/VA 게/ECD 잘/MAG 먹/VV 고/ECE 있/VXV 습니다/EFN",
             "연인/NNG 크림/NNG 새우/NNG 만두/NNG 콜라/NNG 사이즈/NNG 선택/NNG 맛있/VA 었/EPT 는데/ECD 크림/NNG 새우/NNG 가/JKS 너무/MAG 딱/MAG 딱하/VA 어서/ECD 많이/MAG 남기/VV 었/EPT 어요/EFN"]

# 단어의 타입

wordType = set()

for sentence in Sentences:
  # print(sentence)
  words = sentence.split(" ")
  # print(words)
  for word in words:
    wordType.add(word)
print(wordType)
print(len(wordType))
```

## Co-occurence matrix (words-words matrix)

- CS124(From Languages to Information): https://www.youtube.com/channel/UC_48v322owNVtORXuMeRmpA

```python
import pandas as pd

# 0으로 이루어진 단어 행렬 생성 (초기값)
matrix = pd.DataFrame(0.0, index=wordType, columns=wordType)
print(matrix)
```

```python
# 동시 등장 단어 행렬 생성 함수 만들기

for sentence in Sentences:

  # print(sentence)
  words = sentence.split(" ")
  # print(words)
  for i in range(0,len(words)):
    # print(words[i])
    for j in range(0,len(words)):
      # print(words[j])
      if j==i: # 자기 자신인 경우
        pass
      else:
        matrix[words[i]][words[j]] += 1.0
    
print(matrix)
```

## Pointwise mutual information (PMI)
- Dan Jurafsky: https://web.stanford.edu/~jurafsky/
- CS124(From Languages to Information): https://www.youtube.com/channel/UC_48v322owNVtORXuMeRmpA
- (0:08): https://www.youtube.com/watch?v=TBUpxFw8oIA&list=PLaZQkZp6WhWxIvz74aEvvVc99o7WuOoQ6&index=5
- 6.6 Pointwise Mutual Information (PMI): https://web.stanford.edu/~jurafsky/slp3/6.pdf
- 두 단어의 출현 확률을 기반으로 관계 값을 계산한다.

```python
# 단어 타입 별 빈도 계산하기
# 단어 타입 계산을 위한 초기 사전 설정

wordTypeList = list(wordType)

wordTokenDic = dict()
for eachType in wordTypeList:
  wordTokenDic[eachType]=0

# for key, value in wordTokenDic.items():
#   print(key,value)

totalWordTokens = 0
for i in range(0,len(wordTypeList)):
    for j in range(0,len(wordTypeList)):
      wordTokenDic[wordTypeList[i]] += matrix[wordTypeList[i]][wordTypeList[j]]
      totalWordTokens += matrix[wordTypeList[i]][wordTypeList[j]]

print(len(wordTypeList))
for key, value in wordTokenDic.items():
  print(key,value)

print("totalWordTokens:",totalWordTokens)
```

```python
# Pointwise mutual information (PMI)

import pandas as pd
import math as math

# 0으로 이루어진 단어 행렬 생성 (초기값)
matrixPMI = pd.DataFrame(0.0, index=wordType, columns=wordType)
print(matrixPMI)

for i in range(0,len(wordTypeList)):
    for j in range(0,len(wordTypeList)):
      pij = matrix[wordTypeList[i]][wordTypeList[j]]/totalWordTokens
      pi = wordTokenDic[wordTypeList[i]]/totalWordTokens
      pj = wordTokenDic[wordTypeList[j]]/totalWordTokens
      inner_p = pij/(pi*pj)
      if inner_p==0:
        pmi = 0
      else:
        pmi = math.log2(inner_p)
      matrixPMI[wordTypeList[i]][wordTypeList[j]] = pmi

print(matrixPMI)
```

```python
# 생각하기: Positive pointwise mutual information (PPMI)
matrixPPMI = pd.DataFrame(0.0, index=wordType, columns=wordType)
print(matrixPPMI)

for i in range(0,len(wordTypeList)):
    for j in range(0,len(wordTypeList)):
      pij = matrix[wordTypeList[i]][wordTypeList[j]]/totalWordTokens
      pi = wordTokenDic[wordTypeList[i]]/totalWordTokens
      pj = wordTokenDic[wordTypeList[j]]/totalWordTokens
      inner_p = pij/(pi*pj)
      if inner_p==0:
        ppmi = 0
      else:
        ppmi = max(math.log2(inner_p),0)
      matrixPPMI[wordTypeList[i]][wordTypeList[j]] = ppmi

print(matrixPPMI)
```
